## Ejercicio 1: Gráficas y funciones

Gráfica las siguientes funciones:

<!--  log(n), (log(n)))^2, log(log(n)), n^3 + log(n)^3 -->
- log(n)
- (log(n))^2
- log(log(n))
- n^3 + log(n)^3

```python
# Gráficas log(n), (log(n)))^2, log(log(n)), n^3 + log(n)^3
# Librerías
import numpy as np
import matplotlib.pyplot as plt
# Datos
n = np.arange(1, 100)
log_n = np.log(n)
log_log_n = np.log(log_n)
log_log_log_n = np.log(log_log_n)
n_3 = n**3
log_n_3 = np.log(n_3)
log_log_n_3 = np.log(log_n_3)
```

### Solución
La gráfica de la función log(n) es:

```python
# log(n)
plt.plot(n, log_n, label='log(n)')
plt.legend()
plt.show()
```

La gráfica de la función (log(n))^2 es:

```python
# (log(n))^2
plt.plot(n, log_log_n, label='(log(n))^2')
plt.legend()
plt.show()
```

La gráfica de la función log(log(n)) es:

```python
# log(log(n))
plt.plot(n, log_log_log_n, label='log(log(n))')
plt.legend()
plt.show()
```

La gráfica de la función n^3 + log(n)^3 es:

```python
# n^3 + log(n)^3
plt.plot(n, n_3, label='n^3 + log(n)^3')
plt.legend()
plt.show()
```

## Ejercicio 2: Funciones de orden superior

Para cada uno de las siguientes pares de funciones decir cual es Big O de la función que se representa por el par.

<!-- f(n) = (n^2 - n)/2, g(n) = 6n -->
- f(n) = (n^2 - n)/2 , g(n) = 6n
<!-- f(n) = n +  2*sqrt(n), g(n) = n2 -->
- f(n) = n +  2 $\sqrt n$ , g(n) = n2
<!-- f(n) = n + nlogn, g(n) = nsqrt(n) -->
- f(n) = n + n log n , g(n) = n $\sqrt n$
<!-- f(n) = n^2 +3n +4, g(n) = n^3 -->
- f(n) = n^2 +3n +4 , g(n) = n^3
<!-- f(n) = n + log(n), g(n) = sqrt(n) -->
- f(n) = n + log n , g(n) = $\sqrt n$

```python
# Configuración de la gráficas
n = np.arange(1, 100)
fn1 = (n**2 - n)/2
gn1 = 6*n
fn2 = n + 2*np.sqrt(n)
gn2 = n**2
fn3 = n + n*np.log(n)
gn3 = n*np.sqrt(n)
fn4 = n**2 + 3*n + 4
gn4 = n**3
fn5 = n + np.log(n)
gn5 = np.sqrt(n)
```

### Solución

Pareja 1: f(n) = (n^2 - n)/2 , g(n) = 6n

Gráfica de f(n)

```python
# Gráfica de f(n)
plt.plot(n, fn1, label='f(n) = (n^2 - n)/2')
plt.legend()
plt.show()
```

Gráfica de g(n)

```python
# Gráfica de g(n)
plt.plot(n, gn1, label='g(n) = 6n')
plt.legend()
plt.show()
```

Gráficamente podemos concluir que para la primera pareja la función que mas crece es f(n).

Pareja 2: f(n) = n +  2*sqrt(n), g(n) = n2

Gráfica de f(n)

```python
# Gráfica de f(n)
plt.plot(n, fn2, label='f(n) = n + 2 $\sqrt{n}$')
plt.legend()
plt.show()
```

Gráfica de g(n)

```python
# Gráfica de g(n)
plt.plot(n, gn2, label='g(n) = $n^2$')
plt.legend()
plt.show()
```

Gráficamente podemos concluir que para la segunda pareja la función que mas crece es g(n).

Pareja 3: f(n) = n + n log n , g(n) = n $\sqrt n$

Gráfica de f(n)

```python
# Gráfica de f(n)
plt.plot(n, fn3, label='f(n) = n + n log n')
plt.legend()
plt.show()
```

Gráfica de g(n)

```python
# Gráfica de g(n)
plt.plot(n, gn3, label='g(n) = n $\sqrt{n}$')
plt.legend()
plt.show()
```

Gráficamente podemos concluir que para la tercera pareja la función que mas crece es g(n).

Pareja 4: f(n) = $n^2$ +3n +4, g(n) = $n^3$

Gráfica de f(n)

```python
# Gráfica de f(n)
plt.plot(n, fn4, label='f(n) = $n^2$ + 3n + 4')
plt.legend()
plt.show()
```

Gráfica de g(n)

```python
# Gráfica de g(n)
plt.plot(n, gn4, label='g(n) = $n^3$')
plt.legend()
plt.show()
```

Gráficamente podemos concluir que para la cuarta pareja la función que mas crece es g(n).

Pareja 5: f(n) = n + log(n), g(n) = $\sqrt n$

Gráfica de f(n)

```python
# Gráfica de f(n)
plt.plot(n, fn5, label='f(n) = n + log n')
plt.legend()
plt.show()
```

Gráfica de g(n)

```python
# Gráfica de g(n)
plt.plot(n, gn5, label='g(n) = $\sqrt{n}$')
plt.legend()
plt.show()
```

Gráficamente podemos concluir que para la quinta pareja la función que mas crece es f(n).
